<h1 style="color:blue" align="center">Cross-Validation</h1>

Cross-validation is a technique that is widely used in machine learning to evaluate the performance of a model. The basic idea behind cross-validation is to split the available data into two sets: a training set and a validation set. The model is then trained on the training set and tested on the validation set. This process is repeated several times, with different splits of the data, to obtain a more accurate estimate of the model's performance.

The main advantage of cross-validation is that it allows us to use all the available data for both training and testing the model. This is particularly important when the available dataset is small. In addition, cross-validation allows us to estimate the generalization error of the model, i.e., how well the model is likely to perform on new, unseen data.

## Why cross-validation?

By partitioning the available data into three sets, we drastically reduce the number of samples which can be used for learning the model, and the results can depend on a particular random choice for the pair of (train, validation) sets.

A solution to this problem is a procedure called cross-validation (CV for short). A test set should still be held out for final evaluation, but the validation set is no longer needed when doing CV. In the basic approach, called k-fold CV, the training set is split into k smaller sets

<h2 style="color:blue" align="center">1. K-Folds Cross-Validation</h2>

### How K-folds cross-validation operates?
K-fold cross-validation is the most commonly used type of cross-validation. In K-fold cross-validation, the data is divided into K equal parts, or "folds". The model is then trained on K-1 folds and tested on the remaining fold. This process is repeated K times, with each fold being used exactly once as the validation data. **The results are then averaged to obtain a single performance estimate.**


1: Divide data into K partitions(These partitions will be of equal size).

2: Treat Fold-1 as test fold while K-1 as train folds.

3: Compute score of test-fold.

4: Repeat step 3 for all folds taking another fold as test while remaining as train.

5: Take average of scores of all the folds.

<div class="alert alert-block alert-danger"> 

**NOTE: This code is only for visual explanation**
    </div>

```python
# simulate splitting a dataset of 25 observations into 5 folds
from sklearn.model_selection import KFold
kf = KFold(n_splits=5, shuffle=False).split(range(25))

# print the contents of each training and testing set
print('{} {:^61} {}'.format('Iteration', 'Training set observations', 'Testing set observations'))
for iteration, data in enumerate(kf, start=1):
    print('{:^9} {} {:^25}'.format(iteration, data[0], str(data[1])))
```

```python
# Difference between shuffling and not shuffling data
kf = KFold(n_splits=5, shuffle=True).split(range(25)) 

# print the contents of each training and testing set
print('{} {:^61} {}'.format('Iteration', 'Training set observations', 'Testing set observations'))
for iteration, data in enumerate(kf, start=1):
    print('{:^9} {} {:^25}'.format(iteration, data[0], str(data[1])))
```

- Dataset contains **25 observations** (numbered 0 through 24)
- 5-fold cross-validation, thus it runs for **5 iterations**
- For each iteration, every observation is either in the training set or the testing set, **but not both**
- Every observation is in the testing set **exactly once**

```python
from sklearn.datasets import load_iris

iris=load_iris()

X = iris.data[:, :6]  
y = iris.target
```

### Cros_val_score:

Evaluate a score by cross-validation.

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
# All the evaluation metrics
import sklearn
sklearn.metrics.SCORERS.keys()
```

```python
# Import the sklearn library
from sklearn.model_selection import cross_val_score

#fit and evaluate the model.
scores = cross_val_score(knn, X, y, cv=10, scoring='accuracy')
print(scores)
```

```python
#Mean Accuracy
print(scores.mean())
```

```python
scores = cross_val_score(knn, X, y, cv=10, scoring='precision_weighted')
print(scores)
```

```python
#Mean Precision
print(scores.mean())
```

```python
scores = cross_val_score(knn, X, y, cv=10, scoring='recall_weighted')
print(scores)
```

```python
#Mean Recall
print(scores.mean())
```

#### Why 10?!

A value of k=10 is very common in the field of applied machine learning, and is recommend if you are struggling to choose a value for your dataset.

### Cross_validate:

Used to run cross-validation on multiple metrics and also to return train scores, fit times and score times.

```python
from sklearn.model_selection import cross_validate
scores = cross_validate(knn, X, y, cv=10, scoring=['accuracy','precision_weighted','recall_weighted'])
print(scores)
```

```python
print(scores['test_accuracy'].mean())
print(scores['test_precision_weighted'].mean())
print(scores['test_recall_weighted'].mean())
```

`cross_val_score` function is used to evaluate the performance of a model by splitting the data into k-folds, training the model on k-1 folds and testing it on the remaining fold. It returns the evaluation scores for each fold and the mean score across all folds.

On the other hand, `cross_validate` function is used to perform a more comprehensive evaluation of a model by computing multiple evaluation metrics and estimating the fit and score times. It returns a dictionary of scores and timing information for each run.

In summary, `cross_val_score` is a simpler and faster function used to compute a single score for model evaluation, while `cross_validate` provides a more detailed evaluation with multiple scores and timing information.

#### Pros:

1- The model has low bias

2- Low time complexity

3- The entire dataset is utilized for both training and validation.
#### Cons:

1- Not suitable for an imbalanced dataset.

<h2 style="color:blue" align="center">2. Stratified K-fold Cross-Validation</h2>

Stratified K-fold cross-validation is a variation of K-fold cross-validation that is used when the data is imbalanced. In this case, the data is divided into K folds such that each fold contains roughly the same proportion of the different classes as the original dataset. This ensures that each fold is representative of the overall distribution of the data.

### Why stratified K-Fold?

K-Fold does not take class imbalancement into consideration. Therefore, in classifications tasks with imbalanced class distributions, we should prefer StratifiedKFold over KFold.

### How Stratified K-folds cross-validation operates?


Lets say class 0 includes 4 instances, whereas class 1 includes 12 instances .. 

The ratio of class 0 to class 1 is 1/3. If we set k=4, then the test sets include three data points from class 1 and one data point from class 0. Thus, training sets include three data points from class 0 and nine data points from class 1.

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import *
import numpy as np

# Initialize the Stratified K-fold Cross-validator with 5 splits
sk=StratifiedKFold(n_splits=5)
```

```python
# Initialize the array to store the accuracy scores
accuracy_scores = []
  
# Perform cross-validation
for train_index, test_index in sk.split(X, y):
    # Split the data into training and test sets
    x_train_fold, x_test_fold = X[train_index], X[test_index]
    y_train_fold, y_test_fold = y[train_index], y[test_index]
    
    # Fit the model to the training data
    knn.fit(x_train_fold, y_train_fold)
    
    # Make predictions on the test data
    y_pred = knn.predict(x_test_fold)
    
    # Calculate the accuracy score and append it to the list
    accuracy_scores.append(accuracy_score(y_test_fold, y_pred))
    
```

```python
# Print the accuracy scores for each fold
print("Accuracy scores for each fold: ", accuracy_scores)

# Calculate the mean accuracy scores
print("Mean accuracy score: ", np.mean(accuracy_scores))
```

#### Pros:
1- Works well for an imbalanced dataset.

#### Cons:
1- Now suitable for time series dataset.

<h2 style="color:blue" align="center">3. Leave one out Cross-validation</h2>

Leave-One-Out cross-validation (LOOCV) is a special case of K-fold cross-validation where K is equal to the number of samples in the dataset **(`LeaveOneOut()` = `KFold(n_splits=n)`)**. In LOOCV, the model is trained on all but one sample and tested on the remaining sample. This process is repeated for each sample in the dataset. LOOCV is computationally expensive, but can provide a more accurate estimate of the model's performance when the dataset is small.


```python
from sklearn.model_selection import LeaveOneOut

lst_accu_stratified = []
loo = LeaveOneOut()
for train_index, test_index in loo.split(X):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    print(X_train, X_test, y_train, y_test)
```

<h2 style="color:blue" align="center">4. Hold-out Cross-validation</h2>

Hold-out cross-validation is a simple approach where the data is divided into a training set and a testing set, with the testing set being used only once to evaluate the model. This approach is less computationally expensive than K-fold cross-validation, but may not provide as accurate an estimate of the model's performance.

**Hold-out Cross-validation is the same as the train-test split.** The Hold-out method involves randomly dividing the dataset into two subsets, one for training and one for testing.

<h2 style="color:blue" align="center">5. Bootsrapping Cross-Validation</h2>

Bootstrap cross-validation is a resampling method that involves randomly sampling subsets of the original dataset with replacement to create new training sets for each iteration. In each iteration, a subset of the original dataset is sampled with replacement to create a new training set and the remaining data is used as the test set. This process is repeated multiple times, and the performance of the model is evaluated by averaging the performance across all iterations.

<img src="./images/bootstrap_concept.png">



```python
from sklearn.utils import resample
```

```python
# Set the number of bootstrap iterations
n_iterations = 10

# Set the size of each bootstrap sample
n_size = int(len(X) * 0.50)

# Run the bootstrap cross-validation
scores = list()
for i in range(n_iterations):
    
    # Select a random sample with replacement
    X_sample, y_sample = resample(X, y, n_samples=n_size)
    
    # Fit the model to the sample data
    knn.fit(X_sample, y_sample)
    
    # Evaluate the model on the test data
    score = knn.score(X_test, y_test)
    
    # Store the score
    scores.append(score)
```

```python
# Calculate the mean score
mean_score = np.mean(scores)

# Print the mean score
print("Mean Score: {:.2f}".format(mean_score))
```
